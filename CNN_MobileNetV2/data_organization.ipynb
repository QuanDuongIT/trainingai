```python
import os
import json
import shutil

# Kết nối Google Drive
drive.flush_and_unmount()  # Ngắt kết nối
drive.mount('/content/drive')  # Kết nối lại

# Đường dẫn đến thư mục chứa bộ dữ liệu
source_dir = "drive/MyDrive/CNN_MobileNetV2/oxford_flowers_102/dataset"
target_dir = "drive/MyDrive/CNN_MobileNetV2/jpg"

# Đọc tệp cat_to_name.json để lấy tên các loài hoa
with open("/content/drive/MyDrive/CNN_MobileNetV2/oxford_flowers_102/cat_to_name.json", "r") as f:
    cat_to_name = json.load(f)

# Tạo thư mục đích nếu chưa tồn tại
if not os.path.exists(target_dir):
    os.makedirs(target_dir)

# Hàm để di chuyển hình ảnh từ thư mục train, valid và test
def organize_images(source_subdir):
    for label_folder in os.listdir(source_subdir):
        label_folder_path = os.path.join(source_subdir, label_folder)
        if os.path.isdir(label_folder_path):
            # Lấy tên loài hoa từ cat_to_name
            flower_name = cat_to_name.get(label_folder, label_folder)  # Sử dụng label nếu không tìm thấy
            flower_dir = os.path.join(target_dir, flower_name)

            if not os.path.exists(flower_dir):
                os.makedirs(flower_dir)

            # print(f"Đang tổ chức thư mục: {label_folder}")
            for image_file in os.listdir(label_folder_path):
                image_file_path = os.path.join(label_folder_path, image_file)
                if image_file.endswith('.jpg') or image_file.endswith('.jpeg') or image_file.endswith('.png'):
                    # print(f"Di chuyển tệp: {image_file}")
                    try:
                        shutil.move(image_file_path, flower_dir)
                    except Exception as e:
                        print(f"Không thể di chuyển tệp {image_file}: {e}")

# Tổ chức dữ liệu từ các thư mục train, valid và test
organize_images(os.path.join(source_dir, 'train'))
organize_images(os.path.join(source_dir, 'valid'))
organize_images(os.path.join(source_dir, 'test'))

print("Dữ liệu đã được tổ chức thành công!")
```
